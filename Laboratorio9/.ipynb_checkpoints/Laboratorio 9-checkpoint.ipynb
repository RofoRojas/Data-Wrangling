## Pregunta 3
### Funcion Normalizadora

```python
def word_norm(text):
  text = text.str.lower()
  text = text.replace('.', '')
  text = text.replace(',', '')
  text = text.replace('!', '')
  text = text.replace(to_replace =r'!', value='',regex=True)
  text = text.replace('á', 'a')
  text = text.replace('é', 'e')
  text = text.replace('í', 'i')
  text = text.replace('ó', 'o')
  text = text.replace('ú', 'u')
  text = text.replace('?', '')
  text = text.replace('!', '')
  text = text.replace(to_replace =r'-', value='',regex=True)
  text = text.replace(')', '')
  text = text.replace('(', '')
  text = text.replace(':', '')
  text = text.replace('/', '')
  return(text)
```

```python
data[['text']] = word_norm(data['text'])
data.head()
```

### Lematizacion

```python
def Lemmatization(text):
    lemmatizer = WordNetLemmatizer() 
    tokeniced_text = nltk.word_tokenize(sentence)
    lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in tokeniced_text])
    return(lemmatized_output)
    
    
```

```python
data['lem_text'] = word_norm(data['text'])
data[['text','lem_text']].head(10)
```

## Pregunta 6 </b>
### Sentiment Analysize

```python
def sentiment_analysis(text):
    polarity = []
    subjectivity = []
    for tw in text:
        tw = TextBlob(tw)
        ann = tw.sentiment
        polarity.append(ann[0])
        subjectivity.append(ann[1])
    return polarity,subjectivity
```

```python
data['Polarity'], data['Subjectivity'] = sentiment_analysis(data.text)
data['Lem_Polarity'], data['Lem_Subjectivity'] = sentiment_analysis(data.lem_text)
```

```python
Pol_prom, Sub_prom = data.Polarity.mean(), data.Subjectivity.mean()
Lem_Pol_prom, Lem_Sub_prom = data.Lem_Polarity.mean(), data.Lem_Subjectivity.mean()
```
